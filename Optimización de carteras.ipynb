La lista de códigos ISIN (International Securities Identification Number) se genera como una cadena de texto en la que cada código ISIN está entre comillas y separado por comas. Para crearla, debes incluir los códigos ISIN válidos de los activos financieros que estás analizando.

```python
ListaISIN = "'IE00BNQ4RL64','LU0570466788','LU0945454980'"
```

Este código realiza una consulta en Google BigQuery para obtener datos de valor liquidativo (NAV) de activos financieros identificados por códigos ISIN

```python
# Proyecto en el que se ejecuta.
project = 'ironia-data'

#Creacion  del cliente

client = bigquery.Client(project=project, credentials=credentials)
sql_str = "SELECT * FROM( SELECT date,isin,nav FROM ironia-data.ironia.navs WHERE date BETWEEN '2020-01-01' AND CURRENT_DATE()) PIVOT (MAX(nav) FOR isin IN (" + ListaISIN + ")) ORDER BY Date ASC"
```

```python
# Descarga Indices de cada fondo
indices = ['GII','SYBA.DE','^IRX']
start = '2020-01-01'
end = '2025-04-10'

datos_indices = yf.download(indices, start=start, end=end, auto_adjust=True)['Close']

# Limpieza indices
mno.matrix(datos_indices)
```

```python
datos_indices.dropna(axis=0, how="any", inplace=True)
mno.matrix(datos_indices)
```

```python
# Factores de Fama y French semanales
file_path = 'https://mba.tuck.dartmouth.edu/pages/faculty/ken.french/ftp/F-F_Research_Data_Factors_weekly_CSV.zip'

ff_factores = pd.read_csv(file_path, skiprows=4, index_col=0, parse_dates=True)


ff_factores = ff_factores.dropna()
ff_factores.index = pd.to_datetime(ff_factores.index, format='%Y%m%d')

ff_factores = ff_factores / 100
datos_ff_factores = ff_factores.loc['2020-01-01':'2025-04-10']
datos_ff_factores
```

```python
# Pasar datos de diario a semanal para evitar reducir el ruido de los datos
## Wed para suavizar el sesgo del inicio y fin de semana
## .first() es para coger el primer dato disponible
datos_fondos.index = pd.to_datetime(datos_fondos.index)
datos_fondos_sem = datos_fondos.resample("W-WED").first()
datos_indices_sem = datos_indices.resample("W-WED").first()
datos_ff_factores_sem = datos_ff_factores.resample("W-WED").first()

datos_fondos_sem , datos_indices_sem , datos_ff_factores_sem
```

```python
# Limpiamos los datos de precios semanales.
## Filtrar y contar los ceros que hay. Pasarlo a NaN y limpiarlo
mask = datos_ff_factores_sem == 0 # True/False si el dato es igual a 0.
datos_ff_factores_sem[mask] = np.nan  # 0 = NaN, para poder limpiar
mno.matrix(datos_ff_factores_sem)
```

```python
datos_ff_factores_sem.dropna(axis=0, how="any", inplace=True)
mno.matrix(datos_ff_factores_sem)
```

```python
# Eliminar ceros y NaN antes de aplicar log (Para previnir posibles errores)
datos_fondos_sem = datos_fondos_sem.replace(0, np.nan).dropna()
datos_indices_sem = datos_indices_sem.replace(0, np.nan).dropna()
datos_ff_factores_sem = datos_ff_factores_sem.replace(0, np.nan).dropna()

# Evitar datos cercanos a 0 (Prevenir errores numericos al hacer retronos log)
datos_fondos_sem = datos_fondos_sem.clip(lower=1e-8)
datos_indices_sem = datos_indices_sem.clip(lower=1e-8)
datos_ff_factores_sem = datos_ff_factores_sem.clip(lower=1e-8)

# Calcular retornos
retornos_datos_fondos_sem = np.log(datos_fondos_sem).diff().dropna()
retornos_datos_indices_sem = np.log(datos_indices_sem).diff().dropna()
retornos_datos_ff_factores_sem = np.log(datos_ff_factores_sem).diff().dropna()
```

```python
rentabilidad_fondos = retornos_datos_fondos_sem.mean()  # Rentabilidad Fondos
volatilidad_fondos = retornos_datos_fondos_sem.std()  # Volatilidad Fondos

fig, axes = plt.subplots(1, 2, figsize=(18, 4))

axes[0].bar(rentabilidad_fondos.index, rentabilidad_fondos.values)
axes[0].set_ylabel('Media')
axes[0].set_title('Retornos')

axes[1].bar(volatilidad_fondos.index, volatilidad_fondos.values,color='green')
axes[1].set_ylabel('Desviación Estándar')
axes[1].set_title('Volatilidad')
```

```python
# Metdodo historico: Se calcula del VaR con datos historicos
def var_historico(r, level=5):
    """
    Devuelve el VaR Histórica de una Serie o DataFrame
    """
    if isinstance(r, pd.DataFrame):
        return r.aggregate(var_historico, level=level)

    elif isinstance(r, pd.Series):
        return -np.percentile(r, level)

    else:
        raise TypeError("Esperamos una Serie o DataFrame")
```

```python
var_95_hist = var_historico(retornos_datos_fondos_sem)
var_95_hist
```

```python
# Gráfica para var_95_hist (Historico)
plt.figure(figsize=(10, 6))
plt.bar(var_95_hist.index, var_95_hist*100, color='blue', alpha=0.5, label='95%')
plt.title('VaR 95% Histórico')
plt.legend()
```

```python
def var_gaussian(r, level=5):
    """
    Devuelve el VaR Gaussiano Paramétrico de una Serie o DataFrame
    """
    if isinstance(r, pd.DataFrame):
        return r.aggregate(var_gaussian, level=level)

    elif isinstance(r, pd.Series):
        # calcula el valor Z asumiendo que es una normal
        z = norm.ppf(level/100)
        return -(r.mean() + z*r.std())

    else:
        raise TypeError("Esperamos una Serie o DataFrame")
```

```python
var_95_gauss = var_gaussian(retornos_datos_fondos_sem)
var_95_gauss
```

```python
# Gráfica para var_95_gauss (Parametrico)
plt.figure(figsize=(10, 6))
plt.bar(var_95_gauss.index, var_95_gauss*100, color='pink', alpha=0.5, label='95%')
plt.title('VaR 95% Parametrico')
plt.legend()
```

```python
retornos_cartera_futuro = retornos_datos_fondos_sem @ pesos
rentabilidad_fondos_futuro = retornos_cartera_futuro.mean()
volatilidad_fondos_futuro = retornos_cartera_futuro.std()

rng = np.random.default_rng(seed=42)

num_simulaciones = 100000
num_periodos = retornos_datos_fondos_sem.shape[0]

retornos_futuros = rng.normal(
    loc=rentabilidad_fondos_futuro,
    scale=volatilidad_fondos_futuro,
    size=(num_periodos, num_simulaciones))
retornos_futuros.shape
```

```python
# Calculo en Introduccion a Python
def calculate_var_cvar(returns, confidence_level=0.95):
    sorted_returns = np.sort(returns)
    var_index = int((1 - confidence_level) * len(sorted_returns))
    var = sorted_returns[var_index]
    cvar = sorted_returns[:var_index].mean()
    return var, cvar

retornos_futuros_var = np.random.randn(275, 3)
results = {}

num_fondos = retornos_futuros_var.shape[1]  # Numero de fondos
for idx in range(num_fondos):
    fondo_returns = retornos_futuros_var[:, idx]
    var, cvar = calculate_var_cvar(fondo_returns, confidence_level=0.95)
    results[f"Fondo {idx+1}"] = {'VaR': var, 'CVaR': cvar}


print(f"Resultados VaR y CVaR por fondo: \n{results}")
```

```python
results_df = pd.DataFrame(results).T
best_portfolio = results_df['CVaR'].idxmax()
print(f'El fondo con el mayor CVaR es: {best_portfolio}')
```

```python
retornos_futuros_df = pd.DataFrame(data=retornos_futuros,
    index=pd.bdate_range(start='2025-04-16', periods=num_periodos))
```

```python
retornos_cartera_futuro.index = pd.to_datetime(retornos_cartera_futuro.index)
retornos_futuros_df.index = pd.to_datetime(retornos_futuros_df.index)

plt.figure(figsize=(10, 5))
plt.plot(retornos_cartera_futuro.cumsum())
plt.axhline(0, color="black")

for idx in range(num_simulaciones):
    plt.plot(pd.concat([retornos_cartera_futuro, retornos_futuros_df[idx]]).cumsum(), alpha=0.5)
plt.title("Rendimiento acumulado de la cartera")
plt.xlabel("Fecha")
plt.ylabel("Rendimiento acumulado")
plt.grid(True)
plt.show()
```

```python
# Metodo Simulacion Montecarlo
print(f"Descripción de los resultados de los retornos proyectados:\nMedia de los retornos por escenario:\n{rentabilidad_fondos}")
print(f"\nDesviación estándar de los retornos por escenario:\n{volatilidad_fondos}")
print(f"\nPercentiles al 95% de confianza de los retornos por escenario:\nSegun Historicos\n{var_95_hist} \nSegun la distribucion normal:\n{var_95_gauss}")

plt.figure(figsize=(12,8))
for idx in range(min(num_simulaciones, 500)):
    retronos_acumulados = np.cumsum(retornos_futuros_df.iloc[:, idx])
    plt.plot(retornos_futuros_df.index, retronos_acumulados, alpha=0.5, label=f"Escenario {idx+1}")

plt.title("Rendimiento acumulado proyectado de la cartera")
plt.xlabel("Fecha")
plt.ylabel("Rendimiento acumulado")
plt.grid(True)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
